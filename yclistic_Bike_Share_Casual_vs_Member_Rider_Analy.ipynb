```python
import pandas as pd
import numpy as np
import warnings
warnings.filterwarnings('ignore')

# Get list of all CSV files
csv_files = sorted([f for f in os.listdir() if f.endswith('.csv')])
print(f"Found {len(csv_files)} CSV files")

# Load all CSVs
dfs = []
for file in csv_files:
    print(f"Loading {file}...", end=' ')
    df = pd.read_csv(file)
    dfs.append(df)
    print(f"✓ ({len(df)} rows)")

# Combine
df = pd.concat(dfs, ignore_index=True)
print(f"\n✓ Combined: {len(df):,} total rows")
```

```python
print(f"Combined dataset shape: {df.shape}")
print(f"Rows: {len(df):,}")
print(f"Columns: {len(df.columns)}")
print(f"\nColumn names:")
print(df.columns.tolist())
print(f"\nFirst few rows:")
df.head()
```

```python
print("Converting timestamps to datetime...")
df['started_at'] = pd.to_datetime(df['started_at'], errors='coerce')
df['ended_at'] = pd.to_datetime(df['ended_at'], errors='coerce')

print(f"✓ Timestamps converted")
print(f"Date range: {df['started_at'].min()} to {df['started_at'].max()}")
```

```python
print("Calculating trip duration...")
df['trip_duration_mins'] = (df['ended_at'] - df['started_at']).dt.total_seconds() / 60

print(f"✓ Trip duration calculated")
print(f"\nDuration statistics (minutes):")
print(df['trip_duration_mins'].describe())
```

```python
print("Extracting temporal features...")
df['year'] = df['started_at'].dt.year
df['month'] = df['started_at'].dt.month
df['month_name'] = df['started_at'].dt.strftime('%B')
df['day_of_week'] = df['started_at'].dt.day_name()
df['hour'] = df['started_at'].dt.hour

print(f"✓ Temporal features extracted")
print(f"\nSample:")
print(df[['started_at', 'month_name', 'day_of_week', 'hour']].head(50))
```

```python
print("Adding season and weekend flag...")

def get_season(month):
    if month in [12, 1, 2]:
        return 'Winter'
    elif month in [3, 4, 5]:
        return 'Spring'
    elif month in [6, 7, 8]:
        return 'Summer'
    else:
        return 'Fall'

df['season'] = df['month'].apply(get_season)
df['is_weekend'] = df['day_of_week'].isin(['Saturday', 'Sunday']).astype(int)

print(f"✓ Season and weekend flag added")
print(f"\nSeason distribution:")
print(df['season'].value_counts())
```

```python
rows_before = len(df)

print(f"Data cleaning: Removing invalid trips...")
print(f"Starting with: {rows_before:,} rows\n")

# Remove trips < 1 minute or > 24 hours
df = df[(df['trip_duration_mins'] >= 1) & (df['trip_duration_mins'] <= 1440)]

# Remove rows with missing critical values
critical_cols = ['started_at', 'ended_at', 'member_casual', 'start_station_name', 'end_station_name']
df = df.dropna(subset=critical_cols)

rows_after = len(df)
rows_removed = rows_before - rows_after

print(f"Rows removed: {rows_removed:,} ({100*rows_removed/rows_before:.2f}%)")
print(f"Final dataset: {rows_after:,} rows")
print(f"\n✓ Data cleaning complete!")
```

```python
print("="*70)
print("FINAL DATASET SUMMARY")
print("="*70)

print(f"\nShape: {df.shape[0]:,} rows, {df.shape[1]} columns")
print(f"Date range: {df['started_at'].min()} to {df['started_at'].max()}")

print(f"\nColumns:")
print(df.columns.tolist())

print(f"\nNull values:")
print(df.isnull().sum())

print(f"\nUser type distribution:")
print(df['member_casual'].value_counts())

print(f"\nBike type distribution:")
print(df['rideable_type'].value_counts())

print("\n✓ Data is clean and ready for analysis!")
```

```python
# ANALYSIS STARTS HERE
import matplotlib.pyplot as plt
import seaborn as sns

print("\n" + "="*70)
print("STARTING DATA ANALYSIS")
print("="*70)

# Quick check
print(f"\nDataset: {len(df):,} rows, {len(df.columns)} columns")
print(f"Memory: {df.memory_usage(deep=True).sum() / 1024 / 1024:.1f} MB")
```

```python
# User type breakdown
user_counts = df['member_casual'].value_counts()
user_pct = df['member_casual'].value_counts(normalize=True) * 100

print("\n" + "="*50)
print("USER TYPE DISTRIBUTION")
print("="*50)
for user_type in ['member', 'casual']:
    if user_type in user_counts.index:
        print(f"{user_type.upper():10} {user_counts[user_type]:>10,} trips ({user_pct[user_type]:>5.1f}%)")
```

```python
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14, 5))

colors = ['#FF6B6B', '#4ECDC4']
user_counts_sorted = user_counts.sort_values(ascending=False)

# Bar chart
ax1.bar(range(len(user_counts_sorted)), user_counts_sorted.values,
        color=colors, edgecolor='black', linewidth=1.5, alpha=0.8)
ax1.set_xticks(range(len(user_counts_sorted)))
ax1.set_xticklabels([x.upper() for x in user_counts_sorted.index], fontsize=11, fontweight='bold')
ax1.set_ylabel('Number of Trips', fontsize=12, fontweight='bold')
ax1.set_title('Trip Count by User Type', fontsize=13, fontweight='bold')
ax1.grid(axis='y', alpha=0.3)
for i, v in enumerate(user_counts_sorted.values):
    ax1.text(i, v + 50000, f'{v:,.0f}', ha='center', fontweight='bold')

# Pie chart
ax2.pie(user_counts_sorted, labels=[x.upper() for x in user_counts_sorted.index],
        autopct='%1.1f%%', colors=colors, startangle=90,
        textprops={'fontsize': 12, 'weight': 'bold'})
ax2.set_title('User Type Proportion', fontsize=13, fontweight='bold')

plt.tight_layout()
plt.show()
```

```python
duration_stats = df.groupby('member_casual')['trip_duration_mins'].agg([
    ('Count', 'count'),
    ('Mean', 'mean'),
    ('Median', 'median'),
    ('Std Dev', 'std'),
    ('Min', 'min'),
    ('Max', 'max')
]).round(2)

print("\n" + "="*70)
print("TRIP DURATION STATISTICS (minutes)")
print("="*70)
print(duration_stats)

casual_mean = duration_stats.loc['casual', 'Mean']
member_mean = duration_stats.loc['member', 'Mean']
diff_pct = (casual_mean / member_mean - 1) * 100

print(f"\n💡 KEY INSIGHT: Casual riders take {diff_pct:.0f}% LONGER trips!")
print(f"   Casual: {casual_mean:.1f} min | Member: {member_mean:.1f} min")
```

```python
fig, axes = plt.subplots(1, 2, figsize=(14, 5))
df_capped = df[df['trip_duration_mins'] <= 80].copy()

# Box plot
df_capped.boxplot(column='trip_duration_mins', by='member_casual', ax=axes[0])
axes[0].set_title('Trip Duration Distribution', fontweight='bold', fontsize=12)
axes[0].set_xlabel('User Type', fontweight='bold')
axes[0].set_ylabel('Duration (minutes)', fontweight='bold')
axes[0].get_figure().suptitle('')
axes[0].set_xticklabels(['Casual', 'Member'])
axes[0].grid(True, alpha=0.3)

# Violin plot
sns.violinplot(data=df_capped, x='member_casual', y='trip_duration_mins',
               ax=axes[1], palette=['#FF6B6B', '#4ECDC4'], inner='box')
axes[1].set_title('Trip Duration Distribution (Violin)', fontweight='bold', fontsize=12)
axes[1].set_xlabel('User Type', fontweight='bold')
axes[1].set_ylabel('Duration (minutes)', fontweight='bold')
axes[1].set_xticklabels(['Casual', 'Member'])
axes[1].grid(True, alpha=0.3, axis='y')

plt.tight_layout()
plt.show()
```

```python
hourly_usage = df.groupby(['hour', 'member_casual']).size().unstack(fill_value=0)

fig, ax = plt.subplots(figsize=(14, 6))

hourly_usage['casual'].plot(ax=ax, label='Casual', linewidth=3, marker='o',
                             color='#FF6B6B', markersize=6, alpha=0.8)
hourly_usage['member'].plot(ax=ax, label='Member', linewidth=3, marker='s',
                             color='#4ECDC4', markersize=6, alpha=0.8)

ax.set_title('Hourly Usage Pattern: Casual vs. Member Riders', fontsize=14, fontweight='bold')
ax.set_xlabel('Hour of Day', fontsize=12, fontweight='bold')
ax.set_ylabel('Number of Trips', fontsize=12, fontweight='bold')
ax.legend(fontsize=11, loc='upper left')
ax.grid(True, alpha=0.3)
ax.set_xticks(range(0, 24, 2))

# Highlight commute hours
ax.axvspan(7, 9, alpha=0.1, color='blue')
ax.axvspan(16, 18, alpha=0.1, color='red')

plt.tight_layout()
plt.show()

print(f"\nMembers peak: {hourly_usage['member'].idxmax()}:00")
print(f"Casual peak: {hourly_usage['casual'].idxmax()}:00")
```

```python
day_order = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']
daily_usage = df.groupby(['day_of_week', 'member_casual']).size().unstack(fill_value=0)
daily_usage = daily_usage.reindex(day_order)

fig, ax = plt.subplots(figsize=(14, 6))

x = np.arange(len(day_order))
width = 0.35

bars1 = ax.bar(x - width/2, daily_usage['casual'], width, label='Casual',
               color='#FF6B6B', alpha=0.8, edgecolor='black')
bars2 = ax.bar(x + width/2, daily_usage['member'], width, label='Member',
               color='#4ECDC4', alpha=0.8, edgecolor='black')

ax.set_title('Daily Usage Pattern: Weekdays vs. Weekends', fontsize=14, fontweight='bold')
ax.set_xlabel('Day of Week', fontsize=12, fontweight='bold')
ax.set_ylabel('Number of Trips', fontsize=12, fontweight='bold')
ax.set_xticks(x)
ax.set_xticklabels(day_order, rotation=45, ha='right')
ax.legend(fontsize=11)
ax.grid(axis='y', alpha=0.3)

# Highlight weekend
ax.axvspan(4.5, 6.5, alpha=0.1, color='orange')

plt.tight_layout()
plt.show()

# Calculate weekday vs weekend
member_weekday = daily_usage.loc[['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday'], 'member'].mean()
member_weekend = daily_usage.loc[['Saturday', 'Sunday'], 'member'].mean()
casual_weekday = daily_usage.loc[['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday'], 'casual'].mean()
casual_weekend = daily_usage.loc[['Saturday', 'Sunday'], 'casual'].mean()

print(f"\nMEMBERS:")
print(f"  Weekday avg: {member_weekday:,.0f} trips/day")
print(f"  Weekend avg: {member_weekend:,.0f} trips/day")
print(f"\nCASUAL:")
print(f"  Weekday avg: {casual_weekday:,.0f} trips/day")
print(f"  Weekend avg: {casual_weekend:,.0f} trips/day")
```

```python
monthly_usage = df.groupby(['month', 'member_casual']).size().unstack(fill_value=0)

fig, ax = plt.subplots(figsize=(14, 6))

monthly_usage['casual'].plot(ax=ax, label='Casual', linewidth=3, marker='o',
                              color='#FF6B6B', markersize=8, alpha=0.8)
monthly_usage['member'].plot(ax=ax, label='Member', linewidth=3, marker='s',
                              color='#4ECDC4', markersize=8, alpha=0.8)

ax.set_title('Monthly Usage Trends: Seasonal Patterns', fontsize=14, fontweight='bold')
ax.set_xlabel('Month', fontsize=12, fontweight='bold')
ax.set_ylabel('Number of Trips', fontsize=12, fontweight='bold')
ax.legend(fontsize=11, loc='upper left')
ax.grid(True, alpha=0.3)
ax.set_xticks(range(1, 13))
ax.set_xticklabels(['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec'])

# Highlight summer
ax.axvspan(5.5, 8.5, alpha=0.1, color='gold')

plt.tight_layout()
plt.show()

print(f"\nSummer peak (Jun-Aug): Highest usage")
print(f"Winter low (Dec-Feb): Lowest usage")
```
